Data description:
The data files train.csv and test.csv contain gray-scale images of hand-drawn digits, from zero through nine.

Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel-value associated with it, indicating the lightness or darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255, inclusive.

The training data set, (train.csv), has 785 columns. The first column, called "label", is the digit that was drawn by the user. The rest of the columns contain the pixel-values of the associated image.

Each row represents the pixel intensity matrix of an image

```python
from sklearn.model_selection import train_test_split

x=np.array(train.drop('label',axis=1).values)
#it does not drop it permanently from the database, to drop permanently use inplace=true arguement
y=np.array(train.loc[:,'label'])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

Learn(t) how to drop columns in pandas and how to do it temporarily. converting dataframes to array. and splitting into a train test set

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

decisionTreeModel=DecisionTreeClassifier().fit(x_train,y_train)
modelPrediction=decisionTreeModel.predict(x_test)
import sklearn
sklearn.metrics.accuracy_score(y_test, modelPrediction) 
```

```python
def accuracy(y_test,y_pred):
    r=sklearn.metrics.accuracy_score(y_test, y_pred)*100 
    print('The accuracy score is '+str(r)+' percent')
```

```python
accuracy(y_test,modelPrediction)
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, modelPrediction)
print(cm)
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True,cmap="YlGnBu")
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=150)
clf.fit(x_train,y_train)
forestPredictions=clf.predict(x_test)
sklearn.metrics.accuracy_score(y_test,forestPredictions)
```

```python
from sklearn.metrics import confusion_matrix 
cmf = confusion_matrix(y_test, forestPredictions)
sns.heatmap(cmf,annot=True,cmap="YlGnBu")
```
